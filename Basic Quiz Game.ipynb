```python
# Basic Quiz Game in Python

score = 0

print("Welcome to the Basic Quiz Game!")
print("--------------------------------")

# Question 1
print("1. What is the capital of Bangladesh?")
answer = input("Your answer: ")

if answer.lower() == "dhaka":
    print("Correct!\n")
    score += 1
else:
    print("Wrong! The correct answer is Dhaka.\n")

# Question 2
print("2. Which number is even?")
print("a) 3   b) 7   c) 10   d) 5")
answer = input("Your answer: ")

if answer.lower() == "c" or answer == "10":
    print("Correct!\n")
    score += 1
else:
    print("Wrong! The correct answer is 10.\n")

# Question 3
print("3. Who is known as the Father of the Computer?")
answer = input("Your answer: ")

if answer.lower() == "charles babbage":
    print("Correct!\n")
    score += 1
else:
    print("Wrong! The correct answer is Charles Babbage.\n")

# Final Score
print("--------------------------------")
print("Quiz Finished!")
print("Your score is:", score, "/ 3")

if score == 3:
    print("Excellent! üéâ")
elif score == 2:
    print("Good job! üëç")
else:
    print("Keep practicing! üí™")
```
